**Q1: Which country has the highest rate of mortality by road traffic injury in 2019?**

```python
df_indicator_name = df["Indicator Name"]
list_indicators = list(df_indicator_name)
```

```python
def search_spesific_indicator(indicator_like, indicator_list):
    tmp = []
    for indicator in indicator_list:
        if indicator_like.lower() in indicator.lower():
            tmp.append(indicator)
    return set(tmp) 
indicator = list(search_spesific_indicator("traffic", list_indicators))
print(indicator)
```

```python
tmp = df[(df['Indicator Name'].isin(indicator))]
df_target_mortality = tmp[['Country Name', 'Country Code', 'Indicator Name', '2019']]
target = ['Mortality caused by road traffic injury (per 100,000 people)']
df_target_mortality = df_target_mortality[df_target_mortality["Indicator Name"].isin(target)]
df_target_mortality
```

```python
answer_1 = df_target_mortality.loc[df_target_mortality['2019'].idxmax()]
answer_1
```

**Dominican Republic** has the highest rate of mortality by road traffic injury in 2019.
